```python
import numpy as np
import matplotlib.pyplot as plt

def plot_16qam():
    # Définir la matrice de symboles exactement comme le TD
    symbols = np.array([
        [13, 9, 8, 12],
        [5, 1, 0, 4],
        [7, 3, 2, 6],
        [15, 11, 10, 14]
    ])
    
    rows, cols = symbols.shape
    
    # Coordonnées des points IQ (maintenir la symétrie)
    I_values = np.linspace(-3, 3, cols)
    Q_values = np.linspace(3, -3, rows)
    
    # Tracé du diagramme IQ
    plt.figure(figsize=(8, 8))
    
    # Tracer les lignes de la grille
    for i in I_values:
        plt.axvline(i, color='gray', linestyle='--', alpha=0.3)
    for q in Q_values:
        plt.axhline(q, color='gray', linestyle='--', alpha=0.3)
    
    # Tracer les axes principaux
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    
    # Tracer les points et les étiquettes
    for i in range(rows):
        for j in range(cols):
            plt.scatter(I_values[j], Q_values[i], color='red', s=100, zorder=5)
            plt.text(I_values[j], Q_values[i], f'{symbols[i, j]}',
                     fontsize=12, ha='center', va='center', color='white',
                     weight='bold', zorder=6)
    
    # Ajuster les limites pour une meilleure visualisation
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.grid(False)
    plt.xlabel('I (In-Phase)', fontsize=14)
    plt.ylabel('Q (Quadrature)', fontsize=14)
    plt.title('Diagramme de constellation 16-QAM', fontsize=16)
    plt.tight_layout()
    plt.show()

# Afficher uniquement le diagramme 16-QAM
plot_16qam()
```
